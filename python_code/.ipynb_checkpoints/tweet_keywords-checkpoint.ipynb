```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("condensed_dow_and_sentiment.csv")
dataset
```

```python
dataset["Tweet_text"] = dataset["Tweet_text"].str.lower()
```

```python
#use filtered_data for tweets within the desired date range
start_date = '2018-01-01 00:00:00+00:00'
end_date = '2018-12-31 00:00:00+00:00'

begin = dataset["Time"] >= start_date
end = dataset["Time"] <= end_date

between_dates = begin & end
filtered_data = dataset.loc[between_dates]
filtered_data
```

```python
market_filtered = filtered_data[filtered_data['Tweet_text'].str.contains('market')]
market_filtered

#add a column thats the high minus the low for total variation for the day = volatility
```
